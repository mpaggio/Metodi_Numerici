    - tol: tolleranza per il test di arresto.
    - (Nel caso del metodo di bisezione e di falsa posizione si considera per il test di arresto l’ampiezza del sottointervallo confrontata con tolx).
    - nmax= numero massimo di iterazioni
    
- In output:
    - lo zero della funzione
    - il numero di iterazioni effettuate
    - una lista contenente tutti gli iterati xk

#### CORDE:
- La funzione `corde.py` deve assumere come input:

    - il coefficiente angolare comune a tutte le rette
    - il nome `fname` della funzione di cui calcolare lo zero
    - tolx: tolleranza per il test di arresto sull'incremento:          
            |x_k+1-x_k|/|x_k+1|<= tolx
    - tolf: tolleranza per il test di arresto sulla funzione
                |abs(f(x_k+1)|<= tolf
    - nmax= numero massimo di iterazioni
    
- In output:
    - lo zero della funzione
    - il numero di iterazioni effettuate
    - una lista contenente tutti gli iterati xk

#### NEWTON:
- La funzione `newton.py` deve assumere come input:
    - l'iterato iniziale x_0.
    - il nome `fname` della funzione di cui calcolare lo zero
    - il nome `fpname` della derivata prima della funzione di cui calcolare lo zero
    - tolx: tolleranza per il test di arresto sull'incremento:          
               |x_k+1-x_k|/|x_k+1|<= tols
    - tolf: tolleranza per il test di arresto sulla funzione    
                |abs(f(x_k+1)|<= tolf
    - nmax= numero massimo di iterazioni
    
- In output:
    - lo zero della funzione
    - il numero di iterazioni effettuate
    - una lista contenente tutti gli iterati xk

#### SECANTI:
- La funzione `secanti.py` deve assumere come input:
    - due iterati iniziali x_1, x_0.
    - il nome `fname` della funzione di cui calcolare lo zero
    - tolx: tolleranza per il test di arresto sull'incremento:         
               |x_k+1-x_k|/|x_k+1|<= tolx
    - tolf: tolleranza per il test di arresto sulla funzione

- In output:

# METODO DI BISEZIONE

```python
# DEFINISCO L'ALGORITMO DI BISEZIONE:
def bisezione(fname, a, b, tolx):
    fa = fname(a)
    fb = fname(b)
    
    # --- Se il seguente controllo restituisce true, allora vuol dire che non è presente una 
    # --- radice della funzione (perchè gli estremi hanno lo stesso segno):
    if sign(fa)*sign(fb) > 0:
        print("Non è possibile applicare il metodo di bisezione!")
        return None, None, None
    
    iter = 0
    vettore_xk = []
    # --- La seguente formula è quella che mi permette di calcolare 
    # --- il NUMERO MASSIMO DI ITERAZIONI da fare:
    maxit = math.ceil(math.log2(b - a)/tolx) - 1
    
    # --- Criterio di arresto della funzione:
    while abs(b - a) > tolx:
        xk = a + (b - a)/2
        vettore_xk.append(xk)
        iter = iter + 1
        fxk = fname(xk)
        
        # --- Se il seguente controllo è vero, vuol dire che
        # --- abbiamo trovato lo zero della funzione nell'intervallo:
        if fxk == 0:
            return xk, iterazione, vettore_xk
        
        if sign(fa)*sign(fxk) > 0:     # --- Devo continuare a lavorare sull'intervallo [xk,b], perchè con a, ha segni concordi
            a = xk
            fa = fxk
        elif sign(fxk)*sign(fb) > 0:   # --- Devo lavorare sull'altro intervallo [a,xk], perchè con b ha segni concordi
            b = xk
            fb = fxk
            
    return xk, iter, vettore_xk
```

```python
import numpy as np
import matplotlib.pyplot as plt

# UTILIZZO DEL METODO DI BISEZIONE:
f1 = lambda x: np.exp(-x)-(x+1)
a = -1.0
b = 2.0
xx = np.linspace(a,b,200)
plt.plot(xx, f1(xx))
plt.title("exp(-x)-(x+1)")

## --- OSSERVAZIONI --- ##
# Si nota subito ad occhio, che il punto di coordinate (0,0) è uno zero della funzione.
# I valori calcolati dal nostro algoritmo dovranno convergere verso tale valore.

alfa = 0
tolx = 1e-12
tolf = 1e-12
zero_b, iter_b, vettore_xk_b = bisezione(f1,a,b,tolx)
print("Lo zero approssimato con il metodo di bisezione è: ", zero_b)
plt.plot(zero_b, f1(zero_b), "ro")
plt.show()
```

# METODO DI REGULA FALSI

```python
# DEFINISCO L'ALGORITMO DI REGULA FALSI:
def regula_falsi(fname, a, b, tolx, tolf, maxit):
    fa = fname(a)
    fb = fname(b)
    
    # --- Se il seguente controllo restituisce true, allora vuol dire che non è 
    # --- presente una radice della funzione (perchè gli estremi hanno lo stesso segno):
    if sign(fa)*sign(fb) > 0:
        print("Non è possibile applicare il metodo di bisezione!")
        return None, None, None
    
    it = 0
    vettore_xk = []
    fxk = 10
    
    # Scrivo la condizione di fine iterazione:
    while it < maxit and abs(b - a) > tolx and abs(fxk) > tolf:
        xk = a - fa*((b - a)/(fb -fa))
        vettore_xk.append(xk)
        it = it + 1
        fxk = fname(xk)
        
        # --- Se il seguente controllo restituisce vero, allora
        # --- è stato trovato lo zero della funzione:
        if fxk == 0:
            return xk, it, vettore_xk
        
        if sign(fa)*sign(fxk) > 0:     # --- Devo continuare a lavorare sull'intervallo [xk,b], perchè con a, ha segni concordi
            a = xk
            fa = fxk
        elif sign(fxk)*sign(fb) > 0:   # --- Devo lavorare sull'altro intervallo [a,xk], perchè con b ha segni concordi
            b = xk
            fb = fxk
            
    return xk, it, vettore_xk
```

# METODO DELLE CORDE

```python
# DEFINISCO L'ALGORITMO DEL METODO DELLE CORDE:
def corde(fname, m, x0, x, tolx, tolf, nmax):
    xk = []
    fx0 = fname(x0)
    fx = fname(x)
    
    x1 = x0 - fx0*(x - x0)/(fx - fx0)
    fx1 = fname(x1)
    xk.append(x1)
    it = 1
    
    # L'errore relativo fra due iterati successivi deve essere minore della tolleranza:
    while it < nmax and abs(fx1) >= tolf and abs(x1 - x0) >= tolx*abs(x1):
        # Per andare al passo successivo, mi salvo i valori del passo precedente come valori iniziali:
        x0 = x1
        fx0 = fname(x0)
        # Calcolo del valore nel passo successivo:
        x1 = x0 - fx0*(x - x0)/(fx - fx0)
        fx1 = fname(x1)
        it = it + 1
        xk.append(x1)
        
    # Controllo di non aver superato i passi massimi concessi:    
    if it == nmax:
        print("Raggiunto il numero massimo di tentativi")
    # Ritorno finale dei valori
    return x1,it,xk
```

# METODO DELLE SECANTI

```python
# DEFINISCO L'ALGORITMO DEL METODO DELLE CORDE:
def secanti(fname, x0, x1, tolx, tolf, nmax):
    xk = []
    fx0 = fname(x0)
    fx1 = fname(x1)
    mk = (fx1 - fx0)/(x1 - x0)
    
    x = x0 - fx0/mk
    fx = fname(x)
    xk.append(x)
    it = 1
    
    # L'errore relativo fra due iterati successivi deve essere minore della tolleranza:
    while it < nmax and abs(fx) >= tolf and abs(x - x0) >= tolx*abs(x):
        # Per andare al passo successivo, mi salvo i valori del passo precedente come valori iniziali:
        x0 = x1
        x1 = x
        fx0 = fname(x0)
        fx1 = fname(x1)
        mk = (fx1 - fx0)/(x1 - x0)
        # Calcolo del valore nel passo successivo:
        x = x0 - fx0/mk
        fx = fname(x)
        it = it + 1
        xk.append(x)
        
    # Controllo di non aver superato i passi massimi concessi:    
    if it == nmax:
        print("Raggiunto il numero massimo di tentativi")
    # Ritorno finale dei valori
    return x,it,xk
```

# METODO DI NEWTON

```python
import sympy as sym
from sympy.utilities.lambdify import lambdify

# DEFINIZIONE DI UNA VARIABILE SIMBOLICA X:
x = sym.symbols('x')

# DEFINISCO LA FUNZIONE:
fs = sym.exp(-x)-(x+1)

# DEFINISCO LA DERIVATA PRIMA SIMBOLICA DELLA FUNZIONE:
# (passandogli la funzione da derivare, la variabile per cui derivare e quante volte derivare)
dfs = sym.diff(fs,x,1)

# TRASFORMO L'ESPRESSIONE SIMBOLICA DELLA DERIVATA PRIMA IN UNA LAMBDA FUNCTION:
# (in questo modo posso usarla nei calcoli numerici)
# --- il 1° argomento rappresenta la variabile simbolica utilizzata.
# --- il 2° argomento rappresenta la funzione che vogliamo trasformare.
# --- il 3° argomento server per specificare che la lambda function può prendere
# --- come argomento un numpy array ed eseguire su tutte le sue componenti.
fp = lambdify(x, dfs, np)

#TRASFORMO L'ESPRESSIONE SIMBOLICA DELLA FUNZIONE DI PARTENZA IN UNA LAMBDA FUNCTION:
# (ora posso usare entrambe le funzioni nei calcoli numerici).
f = lambdify(x,fs,np)

# Stampa della lambda function in maniera simbolica
print(dfs)
```

```python
# DEFINISCO L'ALGORITMO DEL METODO DI NEWTON:

def newton(fname, fpname, x0, tolx, tolf, nmax):
    xk = []
    fx0 = fname(x0)
    
    if fpname(x0) == 0:
        print("Derivata prima nulla in x0")
        return None, None, None
    
    d = fx0/fpname(x0)
    x1 = x0 - d
    
    fx1 = fname(x1)
    xk.append(x1)
    it = 1
    
    # L'errore relativo fra due iterati successivi deve essere minore della tolleranza:
    while it < nmax and abs(fx1) >= tolf and abs(d) >= tolx*abs(x1):
        # Per andare al passo successivo, mi salvo i valori del passo precedente come valori iniziali:
        x0 = x1
        fx0 = fname(x0)
        d = fx0/fpname(x0)
        # Calcolo del valore nel passo successivo:
        x1 = x0 - d
        fx1 = fname(x1)
        it = it + 1
        xk.append(x1)
        
    # Controllo di non aver superato i passi massimi concessi:    
    if it == nmax:
        print("Raggiunto il numero massimo di tentativi")
    # Ritorno finale dei valori
    return x1,it,xk
```
