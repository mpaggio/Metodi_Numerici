## Esercizio 1 
Si consideri l'equazione quadratica $x^2 + 2px - q = 0 $ con $p=10^5$ e $q=10^{-i}$, i=0,..,10

- Si stabilisca se il problema relativo al calcolo della soluzione $x= -p+\sqrt{p^2+q}$ risulta essere ben
      condizionato per tutti i valori di q assegnati;

- Si calcoli la soluzione x dell'equazione mediante la formula risolutiva indicata al punto precedente e si dica se l'algoritmo di calcolo risulta numericamente
stabile  per i valori di q assegnati

- si individui una forma stabile per il calcolo della soluzione e si verifichi se l'algoritmo di calcolo risulta numericamente stabile per i valori di q assegnati.

```python
# CALCOLIAMO L'INDICE DI CONDIZIONAMENTO K:
# x può essere visto come una funzione dipendente da q (f(q)).
    ## --> f'(q) = 1/(2*(sqrt(p^2+q)))
    ## --> K = abs(q*f'(q)/f(q)) = abs([q]*[1/(2*sqrt(p^2+q))]*[1/(-p+sqrt(p^2+q))]) = (moltiplico dentro le quadre per un valore (p+sqrt(p^2+q))/(p+sqrt(p^2+q)) = ... = abs((p+sqrt(p^2+q))/(2*sqrt(p^2+q)))
          ## = lim(q --> 0) K = abs((p+sqrt(p^2))/(2*sqrt(p^2))) = abs((2p)/(2p)) = abs(1) = 1
          ## Quindi il problema è ben condizionato
        
import numpy as np
p = 10.0**5
q = 10.0**(-np.arange(11))
x1 = -p + np.sqrt(p**2+q)
print("Valore delle soluzioni trovate")
print(x1)
print()
print("Valori di q utilizzati")
print(q)

# L'errore è dato dallo spacing, in quanto per valori sempre più piccoli di q, sommarlo a p^2 influenzerà sempre meno la somma, fino a non influenzarle proprio
# Nel caso in cui q sia minore allo spacing, allora -p + sqrt(p^2+q) = -p + sqrt(p^2) = -p + p = 0
# (Per influire nella somma, q deve essere maggiore dello spacing che è dell'ordine di grandezza di 10**(-6))
print()
print("Lo spacing vale: ", np.spacing(p**2))

# In questo caso si verifica che la soluzione si annulla per valori di q che vanno oltre l'ordine di grandezza 10^(-6).

# SOLUZIONE:
# Una formula che non causa il problema legato alla differenza, si ottiene razionalizzando la formula precedente e ottenendo la formula:
x2 = q/(p+np.sqrt(p**2+q))
print()
print("Valore delle soluzioni trovate aggiustando la formula")
print(x2)
```

Si consideri lo sviluppo in serie dell'esponenziale $e^x=1+x+\frac{1}{2!}x^2+\frac{1}{3!}x^3+\cdots =\sum_{k=0}^\infty \frac{x^k}{k!} $. Si calcoli un'approssimazione dell'esponenziale troncando ad N=100 termini la serie, per valori di x che variano da -10 a 10 incluso. Pensare ad un algoritmo più stabile ricordando che $e^{-x}=\frac{1}{e^{|x|}}$.



```python
N = 100
a = -10.0
b = 10.0
x_values = np.linspace(a,b,N)
exp_values = []
alfa_values = np.exp(x_values)
print("Valori di x di cui calcolare l'esponenziale: ")
print(x_values)

# ALGORITMO INSTABILE
for x in x_values:
    r = x
    fattoriale = 1
    exp = 1 + r
    for i in range(2,N+1):
        r = r*x
        fattoriale = fattoriale * i
        exp = exp + r/fattoriale
    exp_values.append(exp)

print()
print("Stampa dei valori dell'esponenziale, calcolati dall'algoritmo iterativo: ")
print(exp_values)

exp_values = np.array(exp_values)
err_rel = np.abs(exp_values - alfa_values)/np.abs(alfa_values)
print()
print("Errore relativo calcolato: ")
print(err_rel)

# --- stampa dei valori
plt.plot(x_values,exp_values,'b-',x_values,alfa_values,'r--')
plt.title('Approssimazione esponenziale con serie di Taylor troncata')
plt.legend(['exp troncato','exp esatto'])
plt.show()
# --- stampa degli errori
plt.plot(x_values,err_rel)
plt.yscale("log")
plt.title("Errore relativo in scala semi-logaritimica")
plt.show()

#Si nota che l'errore relativo risulta più alto nel caso di esponente negativo, perchè lo sviluppo in serie
#presenta termini a segno alterno, e sappiamo che in situazioni particolari la somma tra numeri a segno alterno
#può provocare la cancellazione di cifre significative e produrre quindi un algoritmo instabile

# -------------------------------------------------------------------------------------------------------------

#Per realizzare un aglrotimo stabile si considera il fatto che exp(-x)= 1/exp(x).
#Si calcola quindi lo sviluppo in serie, troncato ad N termini di exp(abs(x)) e poi se ne considera il reciproco.

# ALGORITMO STABILE
exp_values2 = []
for x in x_values:
    if x < 0:
        x = abs(x)
        r = x
        fattoriale = 1
        esponentiale = 1 + r;
        for i in range(2,N+1):
            r=r*x
            fattoriale = fattoriale*i;
            esponentiale = esponentiale + r/fattoriale;
        exp_values2.append(1/esponentiale)
    else: 
        r = x
        fattoriale = 1
        esponentiale = 1 + r;
        for i in range(2,N+1):
            r = r*x
            fattoriale = fattoriale*i;
            esponentiale = esponentiale + r/fattoriale;
        exp_values2.append(esponentiale)

print()
print("Stampa dei valori dell'esponenziale, calcolati dall'algoritmo iterativo: ")
print(exp_values)

exp_values2 = np.array(exp_values2)
err_rel2 = np.abs(exp_values2 - alfa_values)/np.abs(alfa_values)
print()
print("Errore relativo calcolato: ")
print(err_rel2)

# --- stampa dei valori
plt.plot(x_values,exp_values2,'b-',x_values,alfa_values,'r--')
plt.title('Approssimazione esponenziale con serie di Taylor troncata')
plt.legend(['exp troncato','exp esatto'])
plt.show()
# --- stampa degli errori
plt.plot(x_values,err_rel2)
plt.yscale("log")
plt.title("Errore relativo (algoritmo migliorato) in scala semi-logaritimica")
plt.show()
```
