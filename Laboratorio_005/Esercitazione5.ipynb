## Esercizio 1 
Si consideri l'equazione quadratica $x^2 + 2px - q = 0 $ con $p=10^5$ e $q=10^{-i}$, i=0,..,10

- Si stabilisca se il problema relativo al calcolo della soluzione $x= -p+\sqrt{p^2+q}$ risulta essere ben
      condizionato per tutti i valori di q assegnati;

- Si calcoli la soluzione x dell'equazione mediante la formula risolutiva indicata al punto precedente e si dica se l'algoritmo di calcolo risulta numericamente
stabile  per i valori di q assegnati

- si individui una forma stabile per il calcolo della soluzione e si verifichi se l'algoritmo di calcolo risulta numericamente stabile per i valori di q assegnati.

```python

# CALCOLIAMO L'INDICE DI CONDIZIONAMENTO K:
# x può essere visto come una funzione dipendente da q (f(q)).
    ## --> f'(q) = 1/(2*(sqrt(p^2+q)))
    ## --> K = abs(q*f'(q)/f(q)) = abs([q]*[1/(2*sqrt(p^2+q))]*[1/(-p+sqrt(p^2+q))]) = (moltiplico dentro le quadre per un valore (p+sqrt(p^2+q))/(p+sqrt(p^2+q)) = ... = abs((p+sqrt(p^2+q))/(2*sqrt(p^2+q)))
          ## = lim(q --> 0) K = abs((p+sqrt(p^2))/(2*sqrt(p^2))) = abs((2p)/(2p)) = abs(1) = 1
          ## Quindi il problema è ben condizionato
        
import numpy as np
p = 10.0**5
q = 10.0**(-np.arange(11))
x1 = -p + np.sqrt(p**2+q)
print(x1)
print(q)

# L'errore è dato dallo spacing, in quanto per valori sempre più piccoli di q, sommarlo a p^2 influenzerà sempre meno la somma, fino a non influenzarle proprio
# Nel caso in cui q sia minore allo spacing, allora -p + sqrt(p^2+q) = -p + sqrt(p^2) = -p + p = 0
# (Per influire nella somma, q deve essere maggiore dello spacing che è dell'ordine di grandezza di 10**(-6))
print(np.spacing(p**2))

# In questo caso si verifica che la soluzione si annulla per valori di q che vanno oltre l'ordine di grandezza 10^(-6).

# SOLUZIONE:
# Una formula che non causa il problema legato alla differenza, si ottiene razionalizzando la formula precedente e ottenendo la formula:
x2 = q/(p+np.sqrt(p**2+q))
print(x2)
```

pigreco il valore math.pi.

```python
# INIZIALIZZAZIONI
import math
import matplotlib.pyplot as plt
n = 34
b = np.zeros((n,))
p = np.zeros((n,))
s = np.zeros((n,))

# ALGORITMO
b[0]=2
s[0]=1
for i in range(1,n):
    p[i]=b[i-1]*s[i-1]
    b[i]=2*b[i-1]
    s[i]=math.sqrt((1.0-math.sqrt(1-s[i-1]**2))/2.0)
plt.plot(np.arange(1,n),p[1:n])
plt.show()

# CALCOLO DELLO SPACING DI 1:
# Si nota che fino ad un certo valore, sembra stabilizzarsi su 3.14, ovvero il reale valore di PI.
# Dopo un certo valore esplode, fino ad azzerarsi totalmente
print(np.spacing(1))
print(s**2)

# SOLUZIONE
# Ho ottenuto anche in questo caso la soluzione razionalizzando la formula
b[0]=2
s[0]=1
for i in range(1,n):
    p[i]=b[i-1]*s[i-1]
    b[i]=2*b[i-1]
    s[i]=(s[i-1]/2.0)/math.sqrt((1.0+math.sqrt(1-s[i-1]**2))/2.0)
plt.plot(np.arange(1,n),p[1:n])
plt.show()

# CALCOLO DELL'ERRORE RELATIVO:
Errore_relativo = np.abs(p - math.pi)/np.abs(math.pi)
print("Errore relativo --> ", Errore_relativo)
plt.plot(np.arange(1,n), Errore_relativo[1:n]*100)
plt.show()
```

Si calcoli un'approssimazione dell'esponenziale troncando ad N=100 termini la serie, per valori di x che variano da -10 a 10 incluso. Si calcoli  per ogni valore di x l'errore relativo del valore approssimato rispetto al valore esatto, ottenuto mediante la funzione exp della libreria numpy. Cosa si osserva?
```python
N = 100
values_x = np.arange(-10.0,11.0, 1.0)
print(values_x)

fattoriale = 1.0
for x in values_x:
    for i in range(0,N):
        if i != 0.0:
            fattoriale = fattoriale * i
        ex = (x**i)/(fattoriale)
print(ex)
```
